# Global Health & Mortality Prediction System

Exploratory data analysis and model training notebook for Death Rate and Life Expectancy prediction.

```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append(os.path.join(os.getcwd(), '..'))
from train_model import DATA_PATH, generate_sample_dataset, train_and_save_models

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## Load Dataset

```python
if not os.path.exists(DATA_PATH):
    df = generate_sample_dataset()
    os.makedirs(os.path.dirname(DATA_PATH), exist_ok=True)
    df.to_csv(DATA_PATH, index=False)
else:
    df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df.info()
df.describe()
```

## Exploratory Data Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df['Death Rate'], kde=True, ax=axes[0], color='#e74c3c')
axes[0].set_title('Death Rate Distribution')
sns.histplot(df['Life Expectancy'], kde=True, ax=axes[1], color='#00b894')
axes[1].set_title('Life Expectancy Distribution')
plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr = df[numeric_cols].corr()
plt.figure(figsize=(14, 10))
sns.heatmap(corr, cmap='coolwarm', center=0, annot=False)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='GDP per Capita', y='Life Expectancy', alpha=0.5)
plt.title('GDP per Capita vs Life Expectancy')
plt.show()
```
